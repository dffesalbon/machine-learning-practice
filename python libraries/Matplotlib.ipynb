Matplot Lib

```python
import matplotlib.pyplot as plt
```

```python
#import numpy for data for plotting
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, z)
plt.show()
```

```python
#adding axis labels
plt.plot(x, y)
plt.xlabel('Angle')
plt.ylabel('Sine Value')
plt.title('Sine Wave')
plt.show()
```

```python
#parabola
x = np.linspace(-10, 10, 20)
y = x**2
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, 'r+')
plt.show()
```

```python
plt.plot(x, y, 'g.')
plt.show()
```

```python
plt.plot(x, y, 'rx')
plt.show()
```

```python
x = np.linspace(-5, 5, 50)
plt.plot(x, np.sin(x), 'g-')
plt.plot(x, np.cos(x), 'r--')
plt.show()
```

Bar Graph

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
languages = ['English', 'French', 'Latin', 'Russian', 'Japanese']
people = [100, 50, 167, 178, 46]
ax.bar(languages, people)
plt.xlabel('Languages')
plt.ylabel('People Count')
plt.show()
```

Pie Chart

```python
fig1 = plt.figure()
ax = fig1.add_axes([0, 0, 1, 1])
languages = ['English', 'French', 'Latin', 'Russian', 'Japanese']
people = [100, 50, 167, 178, 46]
ax.pie(people, labels=languages, autopct='%1.1f%%')
plt.xlabel('Languages')
plt.ylabel('People Count')
plt.show()
```

Scatter Plot

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0, 0, 1, 1])
ax.scatter(x, y, color='g')
ax.scatter(x, z, color='r')
plt.show()
```

3D Scatter Plot

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x, y, z, cmap='Blues')
plt.show()
```
