```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.rcdefaults()
fig, ax = plt.subplots()

accuracy = [96.30, 98.50, 96.30, 92.60, 95.60, 93.30]
precision = [96.30, 98.50, 96.30, 92.60, 95.60, 93.30]
recall = [96.30, 98.50, 96.30, 92.60, 95.60, 93.30]
f1 = [96.30, 98.50, 96.30, 92.60, 95.60, 93.30]
auc = [99.60, 99.70, 99.70, 95.90, 99.40, 98.30]

# Example data
algorithms = ['Logistic Regression', 'Support Vector Machine', 'K-Nearest Neighbor', 'Decision Tree Classifier', 'Random Forest Classifier', 'Naive Bayes Classifier']
y_pos = algorithms;
df = pd.DataFrame(accuracy, index=algorithms, columns =['algo'])
df = df.sort_values(by=['algo'], ascending=False)
ax.barh(df.index, df['algo'].tolist(), align='center')
ax.set_yticks(df.index)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Percentage')
ax.set_title('Machine Learning Algorithms Accuracy Comparison')

plt.show()
```
